```python
import rioxarray as rio

DEM_PATH = "data/processed/dem_30m_by_province/appalachian_plateaus/appalachian_plateaus_USGS_1_n34w087.tif"

da_dem = rio.open_rasterio(DEM_PATH, masked=True).squeeze()

da_dem
```

```python
from gudhi import CubicalComplex
cc = CubicalComplex(top_dimensional_cells=da_dem)
print(f"Cubical complex is of dimension {cc.dimension()} - {cc.num_simplices()} simplices.")
```

```python
all_cells = cc.all_cells()

print(all_cells.shape)
print(all_cells)
```
